```python
real_locations = {'geometry': []}
esti_locations = {'geometry': []}
for idx in range(0, len(vis_data['real'])):
    try:
        p_real = Point(float(vis_data['real'][idx][1]), float(vis_data['real'][idx][0]))
        p_esti = Point(float(vis_data['est'][idx][1]), float(vis_data['est'][idx][0]))
        real_locations['geometry'].append(p_real)
        esti_locations['geometry'].append(p_esti)
    except:
        pass

real_points = gpd.GeoDataFrame(real_locations, geometry=real_locations['geometry'])
esti_points = gpd.GeoDataFrame(esti_locations, geometry=esti_locations['geometry'])
```
